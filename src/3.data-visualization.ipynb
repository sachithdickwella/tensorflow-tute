```python
plt.plot(x, y)
plt.title('X - Y Title')
plt.xlabel('X Label')
plt.ylabel('Y Label'); # Note the semicolon removes the text output.
```

```python
plt.plot(x, y)

plt.xlim(0, 2)      # X axis limit
plt.ylim(0, 300)    # Y axis limit

plt.title('X - Y Title')
plt.xlabel('X Label')
plt.ylabel('Y Label');
```

```python
plt.figure(figsize=(8, 5))
plt.plot(x, y, color='#79BF4B', marker='D', markersize=12, linestyle='--')
plt.title('X - Y Title')
plt.xlabel('X Label')
plt.ylabel('Y Label');
```

# Seaborn Basics

```python
import seaborn as sns
```

```python
df = pd.read_csv('../notebooks/DATA/heart.csv')
df
```

```python
p = sns.displot(df['age'], kde=True, bins=50, color='red') # KDE = Kernal Density Estimation (Probability Distribution of Data)
p.fig.set_size_inches(10, 5)

```

### Counterplot

```python
df.head()
```

```python
sns.countplot(x='sex', data=df)
```

```python
sns.countplot(x='cp', data=df)
```

```python
sns.countplot(x='cp', data=df, hue='sex')
```

```python
sns.countplot(x='cp', data=df, hue='sex', palette='terrain')
```

```python
sns.countplot(x='cp', data=df, palette='hot') # https://matplotlib.org/stable/tutorials/colors/colormaps.html
```

```python
sns.boxplot(x='sex', y='age', data=df) # X = Categorical Feature, Y = Continous Feature
plt.show()
```

```python
sns.boxplot(x='target', y='thalach', data=df, hue='sex') # Example with outliers
```

```python
sns.scatterplot(x='chol', y='trestbps', data=df, hue='sex', palette='Dark2', size='age')
plt.show()
```
